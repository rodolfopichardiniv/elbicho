sacamos los valores unique de la columna de paises, por temas de complejidad, no sabemos que paises son, por ejemplo DE, NL, etc. Entonces agruparemos paises en los continentes, pero solo ser[a un ejemplo demostrativo, nada 100% atinado por que no sabemos cuales son los paises.

```python
def continente(residence):
  if residence == "DE" or "US" or "MX":
    return "America"
  elif residence == 'NZ'or 'FR'or 'IN'or 'PK'or 'PL'or'PT'or 'CN'or'GR'or'AE'or 'NL'or 'SI'or 'HK'or 'TR'or 'RS'or'PR':
    return "Europa"
  elif residence == 'CA'or 'AT'or 'NG'or 'PH'or 'ES'or'DK'or 'RU'or 'IT'or 'HR'or 'BG'or 'SG'  or 'JP' or 'GB' or 'HN'or 'HU' or 'CZ'or 'AR'or 'DZ':
    return "Africa"
  elif residence == 'BR'or 'IQ'or 'VN'or 'BE'or 'UA'or'MT'or 'CL'or 'RO'or 'IR'or 'CO'or 'MD'or 'KE'or 'LU'or 'JE'or 'TN'or 'MY'or 'EE':
    return "Asia"
  elif residence == 'AU'or 'BO'or'IE'or 'CH':
    return "Oceania"
```

```python
data['company_location'].apply(lambda x: str(x))
```

```python

data["continentes"] = data["company_location"].apply(continente)
data.continentes.values
```
